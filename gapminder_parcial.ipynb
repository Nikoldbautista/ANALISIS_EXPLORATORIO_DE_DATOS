```python
#Instalo gapminder
!pip install gapminder

# Importo la librería necesaria para los gráficos
import plotly.express as px

# Importo el dataset gapminder
from gapminder import gapminder

# El dataset gapminder ya es un DataFrame con información de países
# Filtro los datos para trabajar solo con el año 1972
df = gapminder[gapminder['year'] == 1972]
```

```python
# Creo el gráfico de burbujas donde el eje X es ingreso y el eje Y es salud
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent", text="country", log_x=True, size_max=60, title="Gapminder World 1972")

# Ajusto la posición y tamaño del texto de los países
fig.update_traces(textposition='top center', textfont=dict(size=9))

# Cambio el estilo del gráfico y elimino el título del eje X para no duplicar "INCOME"
fig.update_layout(plot_bgcolor="#f5f5f5", xaxis_title="", yaxis_title="HEALTH", margin=dict(t=100, b=100))

# Personalizo los valores del eje X para que se vean como dinero y bien distribuidos
fig.update_xaxes(tickvals=[1000, 2000, 4000, 8000, 16000, 32000, 64000, 128000], ticktext=["$1 000", "$2 000", "$4 000", "$8 000", "$16 000", "$32 000", "$64 000", "$128 000"])

# Creo la línea horizontal superior que representa los niveles de ingreso
fig.add_shape(type="line", x0=0, x1=1, y0=0.98, y1=0.98, xref="paper", yref="paper", line=dict(color="gray", width=1))

# Creo las líneas verticales punteadas que dividen los niveles en la parte superior
for x in [0.25, 0.5, 0.75]: fig.add_shape(type="line", x0=x, x1=x, y0=0.94, y1=0.98, xref="paper", yref="paper", line=dict(color="gray", width=1, dash="dash"))

# Agrego el texto que indica los niveles de ingreso
fig.add_annotation(x=0, y=1.02, xref="paper", yref="paper", text="INCOME LEVELS ▶", showarrow=False)

# Coloco los nombres de cada nivel distribuidos uniformemente
fig.add_annotation(x=0.125, y=0.985, text="LEVEL 1", showarrow=False, xref="paper", yref="paper")
fig.add_annotation(x=0.375, y=0.985, text="LEVEL 2", showarrow=False, xref="paper", yref="paper")
fig.add_annotation(x=0.625, y=0.985, text="LEVEL 3", showarrow=False, xref="paper", yref="paper")
fig.add_annotation(x=0.875, y=0.985, text="LEVEL 4", showarrow=False, xref="paper", yref="paper")

# Agrego el texto inferior grande que representa el ingreso
fig.add_annotation(x=0.5, y=-0.08, xref="paper", yref="paper", text="INCOME", showarrow=False, font=dict(size=30, color="gray"))

# Agrego el texto inferior izquierdo indicando menor ingreso
fig.add_annotation(x=0.3, y=-0.08, xref="paper", yref="paper", text="← POOR", showarrow=False, font=dict(size=12))

# Agrego el texto inferior derecho indicando mayor ingreso
fig.add_annotation(x=0.7, y=-0.08, xref="paper", yref="paper", text="RICH →", showarrow=False, font=dict(size=12))

# Finalmente muestro el gráfico completo
fig.show()
```
