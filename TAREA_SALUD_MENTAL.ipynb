```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Salud_mental = pd.read_csv('Salud_Mental_20260412.csv')
```

```python
Salud_mental.head(10)
```

```python
# PASO 1: Selecciono las columnas que corresponden a grupos de edad
columnas_edades = Salud_mental.columns[2:]
# Tomo todas las columnas desde la posición 2 en adelante (omite diagnóstico y código)
```

```python
# PASO 2: Creo una copia del DataFrame para trabajar sin modificar el original
proporciones = Salud_mental.copy()
# Esto evita dañar los datos originales
```

```python
# PASO 3: Calcular la proporción por cada grupo de edad
for col in columnas_edades:  # Recorre cada columna de edad

    total = Salud_mental[col].sum()
    # Suma total de casos en ese grupo de edad

    proporciones[col] = Salud_mental[col] / total
    # Divide cada valor entre el total → convierte a proporción


# PASO 4: Crear un diccionario para guardar resultados
resultado = {}
# Aquí guardaremos: edad → (enfermedad, proporción)


# PASO 5: Encontrar la enfermedad con mayor proporción en cada edad
for col in columnas_edades:  # Recorre cada grupo de edad

    idx_max = proporciones[col].idxmax()
    # Encuentra el índice (fila) donde está el valor más alto

    enfermedad = proporciones.loc[idx_max, 'diagnostico_ingreso']
    # Obtiene el nombre de la enfermedad en esa fila

    valor = proporciones.loc[idx_max, col]
    # Obtiene el valor de la proporción más alta

    resultado[col] = (enfermedad, valor)
    # Guarda el resultado en el diccionario


# PASO 6: Imprimir resultados de forma clara
for edad, (enfermedad, valor) in resultado.items():
    # Recorre cada resultado guardado

    print(f"Grupo de edad: {edad}")
    # Muestra el grupo de edad

    print(f"Enfermedad con mayor proporción: {enfermedad}")
    # Muestra la enfermedad dominante

    print(f"Proporción: {valor:.4f}")
    # Muestra la proporción con 4 decimales

    print("-" * 40)
    # Línea separadora para que se vea ordenado


# PASO 7 (OPCIONAL): Convertir resultados en una tabla
df_resultado = pd.DataFrame([
    {"Grupo de edad": edad, "Enfermedad": enf, "Proporción": val}
    for edad, (enf, val) in resultado.items()
])
# Crea un DataFrame con los resultados para visualizar mejor

df_resultado
# Muestra la tabla final
```
